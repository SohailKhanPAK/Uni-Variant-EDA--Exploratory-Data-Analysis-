# 1. Categorical Data

## a. Countplot

```python


# Count values
sex_counts = df['Survived'].value_counts()

# Plot (hue کے بغیر تاکہ مسئلہ نہ ہو)
ax = sns.countplot(x=df['Survived'], hue=df['Survived'], palette='viridis')

# Custom legend using Matplotlib
legend_labels = [f"{category} ({count})" for category, count in sex_counts.items()]
plt.legend(legend_labels, title="Survived Count")

plt.show()
```

```python


# Count values
sex_counts = df['Pclass'].value_counts()

# Plot (hue کے بغیر تاکہ مسئلہ نہ ہو)
ax = sns.countplot(x=df['Pclass'], hue=df['Pclass'], palette='viridis')

# Custom legend using Matplotlib
legend_labels = [f"{category} ({count})" for category, count in sex_counts.items()]
plt.legend(legend_labels, title="Pclass Count")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Count values
sex_counts = df['Sex'].value_counts()

# Plot (hue کے بغیر تاکہ مسئلہ نہ ہو)
ax = sns.countplot(x=df['Sex'], hue=df['Sex'], palette='cubehelix')

# Custom legend using Matplotlib
legend_labels = [f"{category} ({count})" for category, count in sex_counts.items()]
plt.legend(legend_labels, title="Sex Count")

plt.show()
```

## b. PieChart 

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```
